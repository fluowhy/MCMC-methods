```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def h(x, der=False):
    if der:
        return np.cos(x)
    else:
        return np.sin(x)


def g(x, der=False):
    if der:
        return 2*x
    else:
        return x**2


def f(x, der=False):
    if der:
        return np.ones(len(x))*2
    else:
        return 2*x
    
def y(x):
    return h(x)


X = np.linspace(0, 10, 10)
Y = h(g(f(X)))
Y_d = h(g(f(X, der=True), der=True), der=True)
print(Y)
print(Y_d)
```

```python
plt.plot(X, Y)
plt.plot(X, Y_d)
```
